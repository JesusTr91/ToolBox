```python
import pandas as pd

x=pd.read_csv('BR.csv')
union=x[['Hora de Unión']]
salida=x[['Hora de salida']]
union['Valor']=1
salida['Valor']=-1
salida=salida.set_axis(['Hora de Unión', 'Valor'], axis=1)
```

```python
valores=pd.concat([union, salida])
valores=valores.reset_index()
valores=valores[['Hora de Unión', 'Valor']]
valores
```

```python
sort= valores.sort_values(by='Hora de Unión', ascending=True)
sort=sort.reset_index()
br=sort[['Hora de Unión', 'Valor']]
br.head(25)
```

```python
test= br.groupby(['Hora de Unión', 'Valor'], as_index=False)['Valor'].agg('sum')
test['Acumulado']= test['Valor'].cumsum()
test
```

```python
test.plot.line('Hora de Unión', 'Acumulado');
```

```python
union['Acumulado']=union['Valor'].cumsum()
union.plot.line('Hora de Unión', 'Acumulado')
```
